```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get('https://www.policypredoc.org/#opportunities')
soup = BeautifulSoup(response.text, 'lxml')
print(soup)
```

```python
# Cite: https://www.crummy.com/software/BeautifulSoup/bs4/doc/
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# Create a tag
division = soup.find('div')
print(division)
```

```python
division = soup.find('div')
print(division)
```

```python
# Create a result set
head_para = division.find_all(['h5', 'p'])
print(head_para)
```

```python
all_text = []
for item in head_para:
    all_text.append(item.text)
print(all_text)
```
